```python
name = input("Enter your full name: ")
age = int(input("Enter your age: "))
gpa = float(input("Enter your GPA: "))
courses_input = input("Enter courses enrolled (comma-separated): ")

courses = courses_input.split(", ")
num_courses = len(courses)

student_profile = {
    "name": name,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": num_courses
}

print("\n========== STUDENT PROFILE ==========")
print(f"Name:              {student_profile['name']}")
print(f"Age:               {student_profile['age']}")
print(f"GPA:               {student_profile['gpa']}")
print(f"Courses Enrolled:  {', '.join(student_profile['courses'])}")
print(f"Number of Courses: {student_profile['num_courses']}")

is_deans_list = student_profile['gpa'] >= 3.5
is_minor = student_profile['age'] < 18

print(f"Dean's List:       {'Yes' if is_deans_list else 'No'}")
print(f"Status:            {'Minor' if is_minor else 'Adult'}")
print("======================================")

print("\n--- Type Conversion Demo ---")
print(f"Name   → original: str  → value: {student_profile['name']}")
print(f"Age    → converted to: int   → value: {student_profile['age']}")
print(f"GPA    → converted to: float → value: {student_profile['gpa']}")
print(f"Courses → converted to: list → value: {student_profile['courses']}")
print(f"Dean's List → bool → value: {is_deans_list}")
print(f"Profile → dict → keys: {list(student_profile.keys())}")
```
