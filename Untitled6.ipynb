```python
print("===== SIMPLE CALCULATOR =====")

num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, //, %, **): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero!")
    else:
        result = num1 / num2
        print(f"\n{num1} {operator} {num2} = {result:.4f}")
        print(f"Result is: {'Positive' if result > 0 else 'Negative' if result < 0 else 'Zero'}")
elif operator == "//":
    if num2 == 0:
        print("Error: Cannot divide by zero!")
    else:
        result = num1 // num2
        print(f"\n{num1} {operator} {num2} = {result:.4f}")
        print(f"Result is: {'Positive' if result > 0 else 'Negative' if result < 0 else 'Zero'}")
elif operator == "%":
    result = num1 % num2
elif operator == "**":
    result = num1 ** num2
else:
    print("Error: Invalid operator!")

if operator in ["+", "-", "*", "%", "**"]:
    print(f"\n{num1} {operator} {num2} = {result}")
    print(f"Result is: {'Positive' if result > 0 else 'Negative' if result < 0 else 'Zero'}")

print("=============================")
```
